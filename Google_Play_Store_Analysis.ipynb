```python

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv("/content/gdrive/MyDrive/Google Playstore Data.xlsx - googleplaystore.csv")
```

```python
data['App Category'].unique()
```

```python
data[data['App Category'] == '1.9']
```

```python
data.loc[10472] = data.loc[10472].shift()
data['App Name'].loc[10472] = data['App Category'].loc[10472]
data['App Category'].loc[10472] = np.nan
data.loc[10472]
```

```python
data['Rating'].unique()
```

```python
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data['Rating'].dtype
```

```python
data['Reviews'].unique()
```

```python
data[data['Reviews'] =='3.0M']
```

```python
data['Reviews'] = data.Reviews.replace("0.0",0)
data['Reviews'] = data.Reviews.replace("3.0M",3000000.0)
data['Reviews'] = data['Reviews'].astype(float)
data['Reviews'].dtype
```

```python
data['App Size'].unique()
```

```python
data['App Size'] = data['App Size'].replace("Varies with device",np.nan)
data['App Size'] = data['App Size'].str.replace("M","000") # All size values became the kilobyte type.
data['App Size'] = data['App Size'].str.replace("k","")
data['App Size'] = data['App Size'].replace("1,000+",1000)
data['App Size'] =data['App Size'].astype(float)
data['App Size'].dtype
```

```python
data['App Installs'].unique()
```

```python
data['App Installs'] = data['App Installs'].str.replace(",","")
data['App Installs'] = data['App Installs'].str.replace("+","")
data['App Installs'] = data['App Installs'].replace("Free",np.nan)
data['App Installs'] = data['App Installs'].astype(float)
data['App Installs'].dtype
```

```python
data['App Price'].unique()
```

```python
data['App Price'] = data['App Price'].replace("Everyone",np.nan)
data['App Price'] = data['App Price'].astype(str).str.replace("$","").astype(float)
data['App Price'].dtype
```

```python
data['App Last Updated on Play Store'].unique()
```

```python
data['Last Updated'] = pd.to_datetime(data['App Last Updated on Play Store'])
data['Last Updated']
```

```python
data.corr()
```

```python
#correlation map
f, ax = plt.subplots(figsize=(6.4,6.4))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt='.1f', cmap='RdYlBu')
plt.show()
```

```python
data.describe()
```

```python


category_reviews = data.groupby('App Category')['Reviews'].mean().reset_index()
sorted_data = category_reviews.sort_values('Reviews', ascending=False)

plt.figure(figsize=(10,8))
sns.barplot(x='App Category', y='Reviews', data=sorted_data)
plt.xticks(rotation=80)
plt.xlabel("Category")
plt.ylabel("Reviews (in tens of millions)")


plt.title("Category and Reviews")
plt.show()
```

```python
category_installs = data.groupby('App Category')['App Installs'].mean().reset_index()
sorted_data = category_installs.sort_values('App Installs', ascending=False)

plt.figure(figsize=(10,8))
sns.barplot(x='App Category', y='App Installs', data=sorted_data)
plt.xticks(rotation=80)
plt.xlabel("Category")
plt.ylabel("Installs (in billions)")

# Remove the "le7" label from y-axis ticks
ax = plt.gca()
ax.set_yticklabels([f'{int(tick/1e7):.0f}' for tick in ax.get_yticks()])

plt.title("Category and Installs")
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=data, x=("App's Content Rating"))
plt.xticks(rotation=80)
plt.title('Content Rating',color = 'blue',fontsize=15)
plt.show()
```
